```python
import os
import sys
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
from nltk.corpus.reader import CategorizedPlaintextCorpusReader
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd
from IPython.display import display
import numpy as np

directory = 'Training/'
documents = []
for filename in os.listdir(directory):
    f = open(directory + filename)
    lines = f.read()
    documents.append(lines)
    

vectorizer = CountVectorizer(stop_words='english')
bag_of_words = vectorizer.fit_transform(documents)
bag_of_words.todense()

svd = TruncatedSVD(n_components=2)

lsa = svd.fit_transform(bag_of_words)

topic_encoded_df = pd.DataFrame(lsa)
topic_encoded_df['data'] = documents

cols = topic_encoded_df.columns.tolist()[-1:] + topic_encoded_df.columns.tolist()[:-1]

#display(topic_encoded_df.sort_values(0, ascending=False))

dictionary = vectorizer.get_feature_names()

# encoding matrix
encoding_matrix = pd.DataFrame(svd.components_, columns=dictionary).T
# 0 for topic 1
encoding_matrix['abs_0'] = np.abs(encoding_matrix[0])
# show words with highest score for topic 1
print(encoding_matrix.sort_values('abs_0', ascending=False))
encoding_matrix.sort_values('abs_0', ascending=False).to_csv('Features with LSA.csv')


```
